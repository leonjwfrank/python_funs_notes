```python
def  divideBy2(decNumber, n=None):
    """10è¿›åˆ¶è½¬æ¢ä¸º2è¿›åˆ¶ï¼Œé»˜è®¤
    :params  decNumber è¦è½¬æ¢çš„æ•°å­—
    :params n è¦è½¬æ¢ä¸ºçš„è¿›åˆ¶ï¼Œé»˜è®¤ä¸º2"""
    digits = "0123456789ABCDEF"
    if not n:
        n = 2
    remstack = Stack()    # æ ˆæ¥å¤„ç†é€†åºç®—æ³•
    while decNumber > 0:
        rem = decNumber % n       # æ±‚ä½™æ•°
        remstack.push(rem)
        decNumber = decNumber // n    # æ•´æ•°é™¤
    binString = ''
    while not remstack.isEmpty():
        binString = binString + digits[remstack.pop()]   # å–ç›¸åº”è¿›åˆ¶ç»„åˆæˆæ•°å­—
    return binString

print(divideBy2(42, 2))
print(divideBy2(42, 8))
print(divideBy2(42, 16))
```

```python
def infixToPostfix(infixexpr):
    prec = {}
    prec["*"] = 3
    prec["/"] = 3
    prec["+"] = 2
    prec["-"] = 2
    prec["("] = 1
    opStack = Stack()
    postfixList = []
#     tokenList = infixexpr.split()    # è§£æè¡¨è¾¾å¼åˆ°å•è¯åˆ—è¡¨
    tokenList = list(infixexpr) # .split()    # è§£æè¡¨è¾¾å¼åˆ°å•è¯åˆ—è¡¨
    print('tokenList:{}'.format(tokenList))
    words = "ABCDEFGHIJKLMNOPQRSTUVWXYZ" 
    nums = "0123456789"
    for token in tokenList:
        if token in "ABCDEFGHIJKLMNOPQRSTUVWXYZ"  or token in "0123456789":
            postfixList.append(token)        # æ“ä½œæ•°
            print('num postfixList:{}'.format(postfixList))
        elif token == "(":
            opStack.push(token)
            print(' ( postfixList:{}'.format(postfixList))
        elif token == ")":
            topToken = opStack.pop()
            while topToken != "(":
                postfixList.append(topToken)
                topToken = opStack.pop()
                print(' ) postfixList:{}'.format(postfixList))
        else:     # æ“ä½œç¬¦
            while (not opStack.isEmpty()) and (prec[opStack.peek()] >= prec[token]):
                postfixList.append(opStack.pop())
                print('peek postfixList:{}'.format(postfixList))
            opStack.push(token)
        print('sign postfixList:{}'.format(postfixList))
    while not opStack.isEmpty():
        postfixList.append(opStack.pop())    # æ“ä½œç¬¦
    print('last postfixList:{}'.format(postfixList))
    return "".join(postfixList)   # åˆæˆåç¼€è¡¨è¾¾å¼å­—ç¬¦
```

```python
class Stack(object):
    def __init__(self):
        self.items = []
    def isEmpty(self):
        return self.items == []
    def push(self, value):
        self.items.append(value)      # æ­¤æ—¶æ€§èƒ½O(1) æ¯” insert(0, value)  O(n)é«˜
    def pop(self):
        return self.items.pop()      # é»˜è®¤å¼¹å‡ºæ ˆé¡¶ï¼Œæ€§èƒ½é«˜äº pop(n)
    def peek(self):
        return self.items[len(self.items)-1]  # è¿”å›æœ€ä¸Šå±‚æ•°æ®
    def size(self):
        return len(self.items)
    
def transform(decNumber, dn=None):
    """10è¿›åˆ¶è½¬æ¢ä¸ºä»»æ„è¿›åˆ¶(2~36)"""
    digits = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    if not dn:
        dn = 2
    remstack = Stack()    # æ ˆæ¥å¤„ç†é€†åºç®—æ³•
    while decNumber > 0:
        rem = decNumber % dn       # æ±‚ä½™æ•°
        remstack.push(rem)
        decNumber = decNumber // dn    # æ•´æ•°é™¤
    binString = ''
    while not remstack.isEmpty():
        binString = binString + digits[remstack.pop()]   # å–ç›¸åº”è¿›åˆ¶ç»„åˆæˆæ•°å­—
    return binString


def map_index(word, n):
    """å­—ç¬¦ä»£è¡¨äº†10è¿›åˆ¶å¤šå°‘å€¼
    nè¡¨ç¤ºè¯¥å­—ç¬¦å±äºå‡ è¿›åˆ¶"""
    dicmap = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    if word not in dicmap:
        return False
    else:
        value = dicmap.index(word)    # å–å¾—å­—ç¬¦è¡¨ç¤º å¤šå°‘å€¼
        assert value <= n    # åˆ¤æ–­è¯¥å€¼åœ¨è¿›åˆ¶èŒƒå›´å†…ï¼Œå¦åˆ™æŠ¥é”™
        return value
    
def transfToTen(decNumber, n=None):
    """ä»»æ„è¿›åˆ¶(2~36)è½¬æ¢ä¸º10è¿›åˆ¶
    n è¡¨ç¤ºå­—ç¬¦çš„åŸè¿›åˆ¶"""
    dicmap = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"

    if not n:   # å¦‚æœæ²¡æœ‰è¯´æ˜æ˜¯å‡ è¿›åˆ¶çš„ï¼Œé»˜è®¤æŒ‰10è¿›åˆ¶å¤„ç†
        n = 10
        
    remstack = Stack()    # æ ˆæ¥å¤„ç†é€†åºç®—æ³• 
    dec_lis = list(str(decNumber))  # è½¬æ¢ä¸ºå­—ç¬¦ä¸²åˆ—è¡¨
    destNum = 0
    loop_n = 0
    while len(dec_lis) > 0:
        current_top = dec_lis.pop()   # å–å¾—å½“å‰æ ˆé¡¶å­—ç¬¦
        real_value = map_index(current_top, n)  # å–å¾—çš„å­—ç¬¦åœ¨ åŸè¿›åˆ¶ä¸­è¡¨ç¤ºå¤šå°‘å€¼
        sums =  real_value * (n ** loop_n)   # è®¡ç®—è¯¥ä½ç½®è¡¨ç¤ºå¤šå°‘ï¼Œä»¥10è¿›åˆ¶è®¡
        destNum = destNum + sums  # 10è¿›åˆ¶ä¸­çš„æ€»æ•°
        loop_n += 1      # ä¸‹ä¸€å¾ªç¯
        print(f'loop_n:{loop_n}, list now:{dec_lis}')
    print(f'source {n} num:{decNumber}, dest #10 num:{destNum}')
    return destNum


def transfAny(M, N, srcNum):
    """
    2 <= M,N <= 36
    æœ¬æ–¹æ³•å°† srcNum è½¬æ¢ä¸º 10 è¿›åˆ¶ï¼Œç„¶åè½¬æ¢ä¸º ç›®æ ‡è¿›åˆ¶
    :params M, åŸ10 è¿›åˆ¶æ•°, è¡¨ç¤ºåŸæ•°æ®æ˜¯å¤šå°‘è¿›åˆ¶çš„
    :params N, ç›®æ ‡10 è¿›åˆ¶æ•°, è¡¨ç¤ºåŸæ•°æ®å°†è¦è½¬æ¢ä¸ºå¤šå°‘è¿›åˆ¶
    :params srcNum
    """
    src_list = list(str(srcNum))
    ten_value = transfToTen(decNumber = srcNum, n=M) # å°†åŸå­—ç¬¦Mè¿›åˆ¶æ•°ï¼Œè½¬æ¢ä¸º10è¿›åˆ¶    
    dest_value = transform(decNumber = ten_value, dn=N)  # å°†10è¿›åˆ¶è½¬æ¢ä¸º ç›®æ ‡è¿›åˆ¶ N è¿›åˆ¶æ•°
    print(f'src value:{srcNum}, its ten_value:{ten_value}, final value:{dest_value}')
    
    return dest_value
    
    
    
```

```python
# å®ç°å‘é‡åŠ ä¹˜ç­‰æ“ä½œ
class Vector(object):
    
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    
    # çœŸå‡å€¼ï¼Œå¦‚æœå‘é‡æ¨¡ä¸º0ï¼Œè¿”å›false
    def __bool__(self):
        return bool(abs(self))

    # å®ç°å‘é‡åŠ æ³•
    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        return Vector(x, y)
    
    # å®ç°å‘é‡ä¹˜æ³•ï¼Œä¾‹å¦‚r*3
    def __mul__(self, scalar):
        return Vector(self.x*scalar, self.y*scalar)
    
    # è¿”å›å‘é‡çš„æ¨¡
    # hypot()è¿”å›æ¬§å‡ é‡Œå¾·èŒƒæ•° sqrt(x*x + y*y)
    def __abs__(self):
        return hypot(self.x, self.y)
    
    # å®ç°__repr__æ–¹æ³•ï¼Œåœ¨æ§åˆ¶å°æ‰“å°å‘é‡æ—¶ä¼šè¾“å‡ºVector(1, 2)
    # å®ç°__str__ï¼Œä½¿ç”¨str()è¿”å›å­—ç¬¦ä¸²
    def __repr__(self):
        return 'Vector(%r, %r)' % (self.x, self.y)
```
